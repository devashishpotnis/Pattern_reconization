```python
import pandas as pd
# import sklearn
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
# from sklearn import metrics
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier 
```

```python

class classifiers:
    def __init__(self):
        self.data = pd.read_csv('dataset//Train.csv')
        self.y = self.data['price_range']
        self.X = self.data.drop('price_range', axis = 1)
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(self.X, self.y, test_size=0.20, random_state=27)
    def naive_bayes(self):
        self.gnb = GaussianNB()
        self.gnb.fit(self.X_train, self.y_train)
        self.y_pred = self.gnb.predict(self.X_test)
        self.accuracy = accuracy_score(self.y_test,self.y_pred)*100
        print(f"Naive Bayes Accuracy: {self.accuracy}%")
    def SVM(self):
        self.svc_model = SVC(C= .1, kernel='linear', gamma= 1)
        self.svc_model.fit(self.X_train, self.y_train)
        self.pred = self.svc_model.predict(self.X_test)
        self.accuracy = accuracy_score(self.y_test,self.pred)*100
        print(f"SVM Accuracy:{self.accuracy}%")
    def knn(self):         
        self.classifier= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )  
        self.classifier.fit(self.X_train, self.y_train)
        self.y_pred= self.classifier.predict(self.X_test)  
#         print(confusion_matrix(self.y_test, self.y_pred))
        self.accuracy = accuracy_score(self.y_test,self.y_pred)*100
        print(f"KNN Accuracy:{self.accuracy}%")
    def decission_tree(self):
        self.model = DecisionTreeClassifier()
        self.model.fit(self.X_train,self.y_train)
        self.pred = self.model.predict(self.X_test)
        self.accuracy = accuracy_score(self.y_test,self.pred)*100
        print(f'Decission Tree Accuracy:{self.accuracy}%')
    def kmeans(self):
        pass
```

```python
def main():
    obj = classifiers()
    while True:
        print("""
             1. Navie Bayes model
             2. Decision Tree
             3. Support Vector Machine(SVM)
             4. K Nearest Neighbour alogorithm(KNN)
             5. K means 
             0. Exit
              """)
        a = int(input("Enter Model: "))
        if a == 1:
            obj.naive_bayes()
        elif a == 2:
            obj.decission_tree()
        elif a == 3:
            obj.SVM()
        elif a == 4:
            obj.knn()
        elif a == 5:
            obj.kmeans()
        elif a == 0:
            break
        else:
            print("Invalid Input")
```

```python
main()
```
